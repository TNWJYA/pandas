```python
import pandas as ka
import numpy as ii

print(ka.__version__)
print(ii.__version__)
```

```python
n_rows = 365 * 24
n_cols = 2
cols = ['col1', 'col2']

df = ka.DataFrame(ii.random.randint(1, 20, size=(n_rows, n_cols)),
                 columns=cols)

df.index = ka.util.testing.makeDateIndex(n_rows, freq='H')
df
```

```python
df.resample('M')['col1'].sum().to_frame()
```
