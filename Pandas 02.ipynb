```python
import pandas as ka #source code untuk cek version
import numpy as ii

print(ka.__version__)
print(ii.__version__)
```

```python
n_rows = 10 #baris yang terdiri dari 10 baris
n_cols = 5 # kolom yang terdiri dari 5 kolom
cols = tuple ('GHIJK') #label yang akan diberikan untuk kolom
df = ka.DataFrame(ii.random.randint(1, 5, size=(n_rows, n_cols)), #object data frame yang akan diisi oleh random integer dari angka 1-5
                  columns=cols) 
df
```

```python
df[(df ['G'] == 1) | (df[ 'G' ] == 3)] 
#operator logika OR
#untuk menseleksi baris - baris pada kolom A hanya bernilai 1 atau 3 saja
```

```python
df[df['G'].isin([1, 3])]

#penyederhanaan dari operator OR
```

```python
df [~df ['G'].isin([1, 3])]

#baris yang diseleksi yang nilai pada kolom A nya tidak berangka 1 atau 3
```
