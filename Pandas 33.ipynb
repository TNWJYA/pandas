# MEMPERSIAPKAN DATA FRAME

```python
d = {'pemain':['Budi', 'Joni', 'Iwan', 'Budi', 'Budi', 'Iwan', 'Asep', "Joni"],
     'goal':[2, 1, 3, 1, 1, 2, 2, 3]}

df = pd.DataFrame(d)
df
```

# Menampilkan Nilai Kamulatif

```python
df['goal'].cumsum().to_frame()
```

```python
df['Jumlah_goal_kamulatif'] = df['goal'].cumsum().to_frame()
df
```

```python
df['Jumlah_goal_kamulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum().to_frame()
df
```

```python
df['cummax']=df['goal'].cummax()
```

```python
df['cummin']=df['goal'].cummin()
```

```python
df['cumprod']=df['goal'].cumprod()
df
```
